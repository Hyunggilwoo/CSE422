```python
#@title Default title text
import numpy as np
from sklearn.decomposition import PCA
from scipy import stats
import matplotlib.pyplot as plt
```


2. Run this to load genome.



```python
genomes = np.loadtxt("pca-data.txt", dtype=object)
#X = pd.read_csv('pca-data.txt', sep=" ", header=None, index_col=0)
```

3. Run it to ignore the 1st three columns of info containing attributes.


```python
col_mode_j = stats.mode(genomes[:,3:])
```

```python
col_mode_j
```

```python
genomes[:,3:]
```

```python
col_mode_j.mode
```

4. Compute the genomes here.


```python
masks = np.ones_like(genomes[:,3:])

binary = genomes[:,3:] == col_mode_j.mode
X = masks * binary
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)
```

5. Compute the 1st and 2nd PCAs.

```python
population =genomes[:,2]
```

```python
unique_labels = np.unique(population)
```

6. Store the information into dictionary.

```python
X_dict = {}
for label in unique_labels:
    indx = np.where(genomes[:,2] == label)[0]
    X_dict[label] = X_pca[indx, :]

X_dict
```

```python
for label in unique_labels:
    plt.scatter(X_dict[label][:,0], X_dict[label][:,1], label = label)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()

plt.show()
```
